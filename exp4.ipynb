```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
columns = ['theta','I2']
df = pd.read_csv('./csv/pol.csv', sep='\s+', names=columns)
df.head()

I2 = df.I2.to_numpy()
theta = df.theta.to_numpy()
theta_rad = np.radians(theta)

cos2theta = np.cos(theta_rad)**2
print(cos2theta)
```

```python
figure = plt.figure()
plt.scatter(cos2theta,I2, marker='x')
plt.xlabel('$cos^2(\Theta)$')
plt.ylabel('$I_2(lx)$')
plt.title('Relação entre ângulo de intensidade do analisador')
figure.savefig('relacao.png',format='png', dpi=1200)
```

```python
linFit = np.polyfit(cos2theta,I2,1)
Y = linFit[0]*cos2theta + linFit[1]

figure2 = plt.figure()
plt.scatter(cos2theta,I2, marker='x')
plt.plot(cos2theta,Y, '-r')
plt.legend(['Dados Coletados', 'Linearização'])
plt.xlabel('$cos^2(\Theta)$')
plt.ylabel('$I_2(lx)$')
plt.title('Linearização da intensidade do analisador')
figure2.savefig('relacaoLin.png',format='png', dpi=1200)
```
